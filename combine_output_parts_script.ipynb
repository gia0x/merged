```python
from pathlib import Path
import re
import pandas as pd
```

```python

# Run the code below after cloning the repo locally

folder = Path("...") # add path to the folder containing the subparts                                                           
files = sorted(folder.glob("standardized_master_part_*.csv"),
               key=lambda p: int(re.search(r'part_(\d+)\.csv$', p.name).group(1)))
df = pd.concat((pd.read_csv(f, low_memory=False) for f in files), ignore_index=True)

out_path = folder / "combined.csv" # or whichever name you prefer
df.to_csv(out_path, index=False)
print(f"Saved to: {out_path}")
```
